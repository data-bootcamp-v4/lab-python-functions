```python
#1. Define a function named `initialize_inventory` that takes `products` as a parameter. Inside the function, implement the code for initializing the inventory dictionary using a loop and user input.

products = ["t-shirt", "mug", "hat", "book", "keychain"] #initial product list
inventory={}#empty inventory dictionary

def initialize_inventory(products):
    #initialize the inventory dictionary using a loop and user input
    for product in products: #iterate over the product list 
        inventory[f"{product}"] = int(input(f'Enter the quantity of "{product}" in the inventory: '))
    return inventory


```

```python
# Define a function named `get_customer_orders` that takes no parameters. Inside the function,
# implement the code for prompting the user to enter the product names using a loop. The function should return the `customer_orders` set.

customer_order=set()
def get_customer_orders():
    customer_order.add(input("enter a product you want to order between t-shirt, mug, hat, book, keychain "))
    while input("do you want to order another product? (yes/no)") == "yes":
        customer_order.add(input("enter the name of a product you want to order between t-shirt, mug, hat, book, keychain"))
    else:

     return customer_order
    
get_customer_orders()

    
    
```

```python
# 3. Define a function named `update_inventory` that takes `customer_orders` and `inventory` as parameters.
#  Inside the function, implement the code for updating the inventory dictionary based on the customer orders.

    
def update_inventory(customer_order, inventory):
    for product in customer_order:
        if product in inventory:
            inventory[product] -= 1
    return inventory
```

```python
# Define a function named `calculate_order_statistics` that takes `customer_orders` and `products` as parameters. 
# Inside the function, implement the code for calculating the order statistics (total products ordered, and percentage of unique products ordered). 
# The function should return these values.

def calculate_order_statistics(customer_order, products):
    totalorder=len(customer_order)
    uniqueproducts=totalorder/len(products)*100
    order_statistics=(totalorder, uniqueproducts)
    return order_statistics 

```

```python
calculate_order_statistics(customer_order, products)
order_statistics = calculate_order_statistics(customer_order, products)  
print(order_statistics)
```

```python
#Define a function named `print_order_statistics` that takes `order_statistics` 
# as a parameter. Inside the function, implement the code for printing the order statistics.
def print_order_statistics(order_statistics):
    print(f"Total products ordered: {order_statistics[0]}")
    print(f"Percentage of unique products ordered: {order_statistics[1]:.2f}%")
```

```python
print_order_statistics(order_statistics)
```

```python
#Define a function named `print_updated_inventory` that takes `inventory` as a parameter. Inside the function, implement the code for printing the updated inventory.

def print_updated_inventory(inventory):
    for products,quantities in inventory.items():
        print(f"{products}: {quantities}")
print_updated_inventory(inventory)
```
